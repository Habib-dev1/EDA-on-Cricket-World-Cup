```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style = 'whitegrid')
```

```python
df = pd.read_csv('world_cup_1975.csv')
df
```

```python
# 1. Total number of matches played in the dataset.

total_matches = int(df['Countries'].count().sum())
# total_matches
print(f'The total number of matches played: {total_matches}')
# df.shape
```

```python
# 2. Which team has the highest win percentage?

winners = {}
total_matches = len(df['Result'])

for i in df['Result']:
  name = i.split('won')[0].strip() # Added strip() to remove leading/trailing spaces
  winners[name] = winners.get(name, 0) + 1

win_percentages = {}
for team, wins in winners.items():

    perc = (wins / total_matches) * 100
    win_percentages[team] = perc

highest_win_team = max(win_percentages, key=win_percentages.get)
highest_win_percentage_value = win_percentages[highest_win_team]


for team, perc in win_percentages.items():
    print(f"{team} : {perc:.2f}%")

print(f"\n{highest_win_team} : {highest_win_percentage_value:.2f}%.")
```
